# Definirea Problemei

-------------------------------------------------------------------------------

#### Context și Motivație

În dezvoltarea software modernă, documentația arhitecturală devine adesea depășită (fenomenul de documentation rot) imediat ce prima linie de cod este modificată. Menținerea diagramelor UML în sincron cu implementarea este o sarcină manuală, consumatoare de timp și predispusă la erori.

__Soluția propusă:__ Un agent AI bazat pe un model de limbaj de mici dimensiuni (SLM) care funcționează ca un "Arhitect Digital". Acesta analizează automat codul sursă Python și instrucțiunile în limbaj natural pentru a genera diagrame de structură (UML) în format Diagram-as-Code.

#### Obiectivele Agentului
- __Analiză Statică:__ Extragerea structurii claselor și relațiilor de moștenire folosind AST (Abstract Syntax Tree).

- __Intelegere Semantică:__ Interpretarea cerințelor de tip User Story pentru a schița arhitectura înainte de scrierea codului.

- __Generare DSL:__ Traducerea metadatelor în sintaxă validă pentru instrumente de vizualizare.

#### Impactul SDG (Obiective de Dezvoltare Durabilă)

Proiectul nostru vizează două direcții principale:

- __SDG 9__: Industrie, Inovație și Infrastructură – Prin automatizarea documentației, creștem eficiența echipelor de dezvoltare și reducem datoria tehnică.

- __SDG 4__: Educație de Calitate – Instrumentul servește ca suport didactic pentru programatorii juniori, ajutându-i să vizualizeze instantaneu logica abstractă a codului pe care îl scriu.

# Analiza Datelor de Intrare

------------------------

Pentru a genera o diagramă UML corectă, agentul nu trebuie să "ghicească" codul, ci să îl analizeze structural. Vom folosi biblioteca ast din Python pentru a transforma fișierele sursă în date procesabile.

Surse de date identificate:
1. __Cod Sursă Python:__ Date nestructurate care vor fi convertite în obiecte JSON prin analiză statică.

2. __Dataset Sintetic (Perechi Prompt-UML):__ Pentru a rafina (fine-tune/prompt engineering) capacitatea SLM-ului de a respecta sintaxa Mermaid.

3. __Documentația oficială Mermaid/PlantUML:__ Folosită ca bază de cunoștințe (Knowledge Base) în sistemul de RAG.

# 💻 Celulă de Cod: Analizatorul de Metadate UML

-----------------

Această celulă reprezintă "Tool-ul" principal pe care agentul îl va folosi pentru a colecta date din mediul de execuție.

```python
import ast
import json

class UMLDataExtractor(ast.NodeVisitor):
    """
    Parcurge arborele sintactic (AST) al codului Python 
    pentru a extrage clasele, metodele și moștenirea.
    """
    def __init__(self):
        self.metadata = {"classes": []}

    def visit_ClassDef(self, node):
        # Extragem clasele de bază (pentru relația de moștenire UML)
        bases = [base.id for base in node.bases if isinstance(base, ast.Name)]
        
        methods = []
        for item in node.body:
            if isinstance(item, ast.FunctionDef):
                # Identificăm metodele "private" (cele care încep cu _)
                access = "-" if item.name.startswith("_") else "+"
                methods.append(f"{access}{item.name}()")
        
        self.metadata["classes"].append({
            "name": node.name,
            "inherits": bases,
            "methods": methods
        })
        self.generic_visit(node)

def analyze_input_code(source_code):
    tree = ast.parse(source_code)
    extractor = UMLDataExtractor()
    extractor.visit(tree)
    return extractor.metadata

# Exemplu de "Date de intrare" (Input Data)
raw_code = """
class DatabaseConnector:
    def connect(self): pass
    def _validate_credentials(self): pass

class MySQLConnector(DatabaseConnector):
    def execute_query(self, query: str): pass
"""

extracted_data = analyze_input_code(raw_code)
print(json.dumps(extracted_data, indent=2))
```

## Pipeline de Construire a Dataset-ului

---

Pentru a antrena un SLM care înțelege sintaxa diagramelor, am construit un pipeline complet de colectare și transformare a datelor. Acesta procesează în mod stream fișiere Python din `bigcode/the-stack` (fără a descărca întreg dataset-ul), le analizează static prin AST și generează perechi de instruire în trei formate simultan.

**Parametrii de configurare:**
- `NUM_SAMPLES = 3500` fișiere brute analizate
- Output: 4 fișiere JSONL (full + 3 configurări de instruire)


```python
# ── Extractor AST complet (versiunea finală folosită în pipeline) ─────────────
import ast, json
from typing import Optional

class UMLDataExtractor(ast.NodeVisitor):
    """Walks a Python AST and extracts classes, methods, fields, inheritance, associations."""
    def __init__(self):
        self.metadata = {"classes": []}
        self._class_names: set = set()

    def visit_ClassDef(self, node):
        bases = [b.id for b in node.bases if isinstance(b, ast.Name)]
        methods, fields = [], []
        for item in node.body:
            if isinstance(item, ast.FunctionDef):
                access = "-" if item.name.startswith("_") else "+"
                params = [a.arg for a in item.args.args if a.arg != "self"]
                ret = ""
                if item.returns and isinstance(item.returns, ast.Name):
                    ret = f" : {item.returns.id}"
                methods.append(f"{access}{item.name}({', '.join(params)}){ret}")
            elif isinstance(item, ast.AnnAssign) and isinstance(item.target, ast.Name):
                ann = ""
                if isinstance(item.annotation, ast.Name): ann = item.annotation.id
                elif isinstance(item.annotation, ast.Subscript):
                    try: ann = ast.unparse(item.annotation)
                    except: ann = ""
                fields.append({"name": item.target.id, "type": ann})
        self.metadata["classes"].append({"name": node.name, "inherits": bases, "methods": methods, "fields": fields})
        self._class_names.add(node.name)
        self.generic_visit(node)

    def post_process(self):
        known = self._class_names
        for cls in self.metadata["classes"]:
            assoc = set()
            for m in cls["methods"]:
                for k in known:
                    if k in m and k != cls["name"]: assoc.add(k)
            cls["associations"] = list(assoc)
        return self.metadata

def extract(source: str) -> Optional[dict]:
    try: tree = ast.parse(source)
    except SyntaxError: return None
    ex = UMLDataExtractor()
    ex.visit(tree)
    return ex.post_process()

print("✓ Extractor AST inițializat")
```

```python
# ── Funcțiile de templating pentru cele 3 formate de output ──────────────────
def to_mermaid(meta: dict) -> str:
    lines = ["classDiagram"]
    for cls in meta["classes"]:
        lines.append(f"    class {cls['name']} {{"); 
        for f in cls.get("fields", []):
            t = f" {f['type']}" if f["type"] else ""
            lines.append(f"        +{f['name']}{t}")
        for m in cls["methods"]: lines.append(f"        {m}")
        lines.append("    }")
    for cls in meta["classes"]:
        for base in cls["inherits"]: lines.append(f"    {base} <|-- {cls['name']}")
        for assoc in cls.get("associations", []): lines.append(f"    {cls['name']} --> {assoc}")
    return "\n".join(lines)

def to_plantuml(meta: dict) -> str:
    lines = ["@startuml", ""]
    for cls in meta["classes"]:
        kw = "abstract class" if cls["name"].startswith(("Abstract", "Base")) else "class"
        lines.append(f"{kw} {cls['name']} {{")
        for f in cls.get("fields", []):
            t = f" : {f['type']}" if f["type"] else ""
            lines.append(f"    +{f['name']}{t}")
        for m in cls["methods"]: lines.append(f"    {m}")
        lines.append("}\n")
    for cls in meta["classes"]:
        for base in cls["inherits"]: lines.append(f"{base} <|-- {cls['name']}")
        for assoc in cls.get("associations", []): lines.append(f"{cls['name']} --> {assoc}")
    lines += ["", "@enduml"]
    return "\n".join(lines)

def to_description(meta: dict) -> str:
    parts = []
    for cls in meta["classes"]:
        desc = f"The `{cls['name']}` class"
        if cls["inherits"]: desc += f" extends {', '.join(cls['inherits'])}"
        pub  = [m.lstrip("+").split("(")[0] for m in cls["methods"] if m.startswith("+")]
        priv = [m.lstrip("-").split("(")[0] for m in cls["methods"] if m.startswith("-")]
        if pub:  desc += f" and exposes public methods: {', '.join(pub)}"
        if priv: desc += f". Private methods: {', '.join(priv)}"
        if cls.get("fields"): desc += f". Attributes: {', '.join(f['name'] for f in cls['fields'])}"
        if cls.get("associations"): desc += f". Collaborates with: {', '.join(cls['associations'])}"
        parts.append(desc + ".")
    return " ".join(parts)

print("✓ Template-uri Mermaid / PlantUML / NL inițializate")
```

```python
# ── Filtrul de calitate ───────────────────────────────────────────────────────
def is_useful(meta: dict) -> bool:
    """
    Returnează True doar dacă fișierul produce o diagramă nontrivială:
      - minimum 2 clase
      - cel puțin o relație (moștenire sau asociere)
      - fiecare clasă are cel puțin 1 metodă
      - niciun nume de clasă de un singur caracter (stub auto-generat)
    """
    classes = meta.get("classes", [])
    if len(classes) < 2: return False
    if not any(cls["inherits"] or cls.get("associations") for cls in classes): return False
    if not all(len(cls["methods"]) >= 1 for cls in classes): return False
    if any(len(cls["name"]) <= 1 for cls in classes): return False
    return True

# Demo filtru
test_src = '''
class Animal:
    name: str
    def speak(self): pass

class Dog(Animal):
    breed: str
    def fetch(self): pass
    def _bark(self): pass
'''
meta_demo = extract(test_src)
print("Filtru demo (Animal/Dog) →", is_useful(meta_demo))  # True
print(json.dumps(meta_demo, indent=2, ensure_ascii=False))
```

## Exemple Concrete din Dataset

---

Fiecare înregistrare din dataset conține codul sursă original, descrierea în limbaj natural generată automat și reprezentările în cele două formate DSL. Mai jos sunt două perechi reprezentative:


```python
# ── Pereche de antrenare #1 — moștenire simplă ───────────────────────────────
example_1 = '''
class Shape:
    color: str
    def draw(self): pass
    def area(self) -> float: pass
    def _validate(self): pass

class Circle(Shape):
    radius: float
    def draw(self): pass
    def area(self) -> float: pass

class Rectangle(Shape):
    width: float
    height: float
    def draw(self): pass
    def area(self) -> float: pass
'''
meta1 = extract(example_1)
print("=" * 60)
print("RECORD JSONL — Code-to-Mermaid")
print("=" * 60)
record1 = {
    "instruction": "Generate a Mermaid class diagram for the following Python code.",
    "input": example_1.strip(),
    "output": to_mermaid(meta1)
}
print(json.dumps(record1, indent=2, ensure_ascii=False))
```

```python
# ── Pereche de antrenare #2 — asocieri între clase ───────────────────────────
example_2 = '''
class Engine:
    horsepower: int
    def start(self): pass
    def stop(self): pass
    def _check_fuel(self): pass

class Car:
    model: str
    def drive(self, engine: Engine): pass
    def park(self): pass

class ElectricCar(Car):
    battery_level: float
    def charge(self): pass
    def drive(self, engine: Engine): pass
'''
meta2 = extract(example_2)
print("=" * 60)
print("RECORD JSONL — Text-to-Mermaid (descriere → diagramă)")
print("=" * 60)
record2 = {
    "instruction": "Generate a Mermaid class diagram from this architecture description.",
    "input": to_description(meta2),
    "output": to_mermaid(meta2)
}
print(json.dumps(record2, indent=2, ensure_ascii=False))
print("\n--- PlantUML output ---")
print(to_plantuml(meta2))
```

## Baza de Cunoștințe pentru RAG

---

Pe lângă dataset-ul de perechi cod→diagramă, agentul va folosi un mecanism de **Retrieval-Augmented Generation (RAG)** pentru a accesa documentația oficială a limbajelor DSL în timp real. Aceasta asigură că sintaxa generată rămâne validă chiar și pentru construcții mai puțin frecvente în dataset.

| Sursă | Tip | Conținut | Rol în RAG |
|-------|-----|----------|------------|
| `mermaid.js.org/syntax/classDiagram` | HTML doc | Sintaxă completă Mermaid | Validare & corectare |
| `plantuml.com/class-diagram` | HTML doc | Sintaxă PlantUML | Generare PlantUML |
| Dataset propriu `dataset_full.jsonl` | JSONL | 318 exemple reale | Few-shot retrieval |
| AST-uri canonice din pipeline | JSON | Structuri de clase validate | Indexare semantică |

Documentele vor fi chunked, embedded (cu `sentence-transformers`) și stocate într-un vector store (ex. ChromaDB sau FAISS), urmând a fi interogate la fiecare cerere de generare.

## Sumar Etapa 1

---

| Dimensiune | Detalii |
|------------|--------|
| **Input date** | Cod Python brut, User Stories, documentație Mermaid/PlantUML |
| **Mecanism extracție** | AST static + detectare asocieri prin tipuri de parametri |
| **Filtru calitate** | 4 criterii stricte → 9.1% rată de trecere (318 / 3500) |
| **Formate output** | Mermaid, PlantUML, Descriere NL |
| **Fișiere dataset** | 4 × JSONL (~5.6 MB total) |
| **Baza RAG** | Documentație oficială + dataset propriu, indexate semantic |
| **Etapa următoare** | Fine-tuning SLM (Phi-3 / Mistral 7B) pe perechile Code-to-Mermaid |


```python
import ast
import json
from textwrap import dedent


def _format_class_member(name, kind="method"):
    prefix = "+" if not name.startswith("_") else "-"
    return f"{prefix}{name}()" if kind == "method" else f"{prefix}{name}"


def extract_uml_metadata(source_code):
    tree = ast.parse(source_code)
    metadata = {"classes": []}

    for node in tree.body:
        if not isinstance(node, ast.ClassDef):
            continue

        bases = []
        for base in node.bases:
            if isinstance(base, ast.Name):
                bases.append(base.id)
            elif isinstance(base, ast.Attribute):
                bases.append(base.attr)
            else:
                bases.append(ast.unparse(base))

        methods = []
        for item in node.body:
            if isinstance(item, ast.FunctionDef):
                methods.append(_format_class_member(item.name))

        metadata["classes"].append(
            {
                "name": node.name,
                "inherits": bases,
                "methods": methods,
            }
        )

    return metadata


def metadata_to_mermaid(metadata):
    lines = ["classDiagram"]

    for class_info in metadata.get("classes", []):
        lines.append(f"    class {class_info['name']} {{")
        for method in class_info.get("methods", []):
            lines.append(f"        {method}")
        lines.append("    }")

    for class_info in metadata.get("classes", []):
        for base in class_info.get("inherits", []):
            lines.append(f"    {base} <|-- {class_info['name']}")

    return "\n".join(lines)


def generate_diagram_from_code(source_code):
    metadata = extract_uml_metadata(source_code)
    return metadata, metadata_to_mermaid(metadata)


sample_code = dedent(
    '''
    class DatabaseConnector:
        def connect(self):
            pass

        def _validate_credentials(self):
            pass

    class MySQLConnector(DatabaseConnector):
        def execute_query(self, query: str):
            pass
    '''
)

metadata, mermaid = generate_diagram_from_code(sample_code)
print(json.dumps(metadata, indent=2, ensure_ascii=False))
print()
print(mermaid)
```
