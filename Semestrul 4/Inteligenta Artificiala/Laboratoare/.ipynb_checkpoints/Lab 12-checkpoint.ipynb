```python
import networkx as nx
import random
import os

def incarca_retele_gml(director_gml):
    retele = []
    if os.path.exists(director_gml):
        for folder in os.listdir(director_gml):
            cale_folder = os.path.join(director_gml, folder)
            for fisier in os.listdir(cale_folder):
                if fisier.endswith(".gml"):
                    cale_fisier = os.path.join(cale_folder, fisier)
                    G = nx.read_gml(cale_fisier, label='id')
                    retele.append((fisier, G))
    return retele

def genereaza_retele_suplimentare():
    retele = []
    retele.append(("Florentine Families", nx.florentine_families_graph()))
    retele.append(("Davis Southern Women", nx.davis_southern_women_graph()))
    retele.append(("Erdos-Renyi", nx.erdos_renyi_graph(50, 0.1)))
    retele.append(("Barabasi-Albert", nx.barabasi_albert_graph(50, 3)))
    retele.append(("Watts-Strogatz", nx.watts_strogatz_graph(50, 4, 0.1)))
    retele.append(("Connected Caveman", nx.connected_caveman_graph(5, 10)))
    return retele
```

```python
def fitness_modularitate(G, comunitati):
    if len(comunitati) <= 1 or len(comunitati) == len(G.nodes):
        return 0.0
    return nx.algorithms.community.quality.modularity(G, comunitati)

def fitness_acoperire(G, comunitati):
    if len(comunitati) <= 1:
        return 0.0
    return nx.algorithms.community.quality.partition_quality(G, comunitati)[0]

def fitness_performanta(G, comunitati):
    if len(comunitati) <= 1:
        return 0.0
    return nx.algorithms.community.quality.partition_quality(G, comunitati)[1]
```

```python
def detectare_comunitati_louvain(G):
    comunitati = nx.algorithms.community.louvain_communities(G)
    return list(comunitati)
```

```python
def decodifica_cromozom(cromozom, lista_noduri):
    G_temp = nx.Graph()
    G_temp.add_nodes_from(lista_noduri)
    for i in range(len(cromozom)):
        nod_sursa = lista_noduri[i]
        nod_destinatie = cromozom[i]
        G_temp.add_edge(nod_sursa, nod_destinatie)

    comunitati = list(nx.connected_components(G_temp))
    return comunitati

def initializare_populatie(G, dimensiune_populatie, lista_noduri):
    populatie = []
    for individ in range(dimensiune_populatie):
        cromozom = []
        for nod in lista_noduri:
            vecini = list(G.neighbors(nod))
            if vecini:
                cromozom.append(random.choice(vecini))
            else:
                cromozom.append(nod)
        populatie.append(cromozom)
    return populatie

def crossover_uniform(parinte1, parinte2):
    copil1 = []
    copil2 = []
    for g1, g2 in zip(parinte1, parinte2):
        if random.random() < 0.5:
            copil1.append(g1)
            copil2.append(g2)
        else:
            copil1.append(g2)
            copil2.append(g1)
    return copil1, copil2

def mutatie(G, cromozom, rata_mutatie, lista_noduri):
    cromozom_mutat = list(cromozom)
    for i in range(len(cromozom_mutat)):
        if random.random() < rata_mutatie:
            nod = lista_noduri[i]
            vecini = list(G.neighbors(nod))
            if vecini:
                cromozom_mutat[i] = random.choice(vecini)
    return cromozom_mutat
```

```python
def selectie_turneu(populatie, scoruri, dimensiune_turneu=3):
    participanti = random.sample(list(zip(populatie, scoruri)), dimensiune_turneu)
    castigator = max(participanti, key=lambda x: x[1])
    return castigator[0]

def ruleaza_algoritm_evolutiv(G, generatii=50, dim_populatie=100, rata_mutatie=0.05, functie_fitness=fitness_modularitate):
    lista_noduri = list(G.nodes)
    populatie = initializare_populatie(G, dim_populatie, lista_noduri)

    cel_mai_bun_global_cromozom = None
    cel_mai_bun_global_scor = -float('inf')
    cele_mai_bune_comunitati = []

    for generatie in range(generatii):
        scoruri = []
        for cromozom in populatie:
            coms = decodifica_cromozom(cromozom, lista_noduri)
            scor = functie_fitness(G, coms)
            scoruri.append(scor)

            if scor > cel_mai_bun_global_scor:
                cel_mai_bun_global_scor = scor
                cel_mai_bun_global_cromozom = list(cromozom)
                cele_mai_bune_comunitati = coms

        populatie_noua = [cel_mai_bun_global_cromozom]

        while len(populatie_noua) < dim_populatie:
            p1 = selectie_turneu(populatie, scoruri)
            p2 = selectie_turneu(populatie, scoruri)

            c1, c2 = crossover_uniform(p1, p2)
            c1 = mutatie(G, c1, rata_mutatie, lista_noduri)
            c2 = mutatie(G, c2, rata_mutatie, lista_noduri)

            populatie_noua.extend([c1, c2])

        populatie = populatie_noua[:dim_populatie]

    return cel_mai_bun_global_cromozom, cele_mai_bune_comunitati, cel_mai_bun_global_scor
```

```python
director_date = "C:/Laborator AI/Laborator 11/real"
retele_toate = incarca_retele_gml(director_date) + genereaza_retele_suplimentare()

for nume_retea, G in retele_toate:
    G = nx.convert_node_labels_to_integers(G)

    print("-" * 50)
    print(f"Retea: {nume_retea} | Noduri: {len(G.nodes)} | Muchii: {len(G.edges)}")
    print("-" * 50)

    comunitati_predef = detectare_comunitati_louvain(G)
    scor_predef = fitness_modularitate(G, comunitati_predef)
    print("Metoda Predefinita (Louvain):")
    print(f" -> Comunitati identificate: {len(comunitati_predef)}")
    print(f" -> Performanta (Modularitate): {scor_predef:.4f}")

    cel_mai_bun_cromozom, comunitati_evol, scor_evol = ruleaza_algoritm_evolutiv(
        G,
        generatii=30,
        dim_populatie=50,
        rata_mutatie=0.05,
        functie_fitness=fitness_modularitate
    )

    print("\nMetoda Algoritm Evolutiv:")
    print(f" -> Comunitati identificate: {len(comunitati_evol)}")
    print(f" -> Performanta (Modularitate): {scor_evol:.4f}")

    rezultat_apartenenta = {}
    for id_comunitate, comunitate in enumerate(comunitati_evol):
        for nod in comunitate:
            rezultat_apartenenta[nod] = id_comunitate

    noduri_exemplu = list(rezultat_apartenenta.keys())[:5]
    print(f" -> Exemplu apartenenta noduri: { {n: rezultat_apartenenta[n] for n in noduri_exemplu} }")
    print("\n")
```
