```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot
from numpy import linspace
```

```python
m = 165
s = 10
data = norm.rvs(loc=m, scale=s, size=1000)
```

```python
x = linspace(130, 210, 1000)
plot(x,norm.pdf(x, loc = m, scale = s), '-r')
hist(data, bins=16, density=True, range=(130, 210))
xticks(range(130, 210, 5), fontsize=10)
grid()
show()
```

```python
from scipy.stats import expon, uniform
from numpy import mean, std

def getRandomTime():
    Y = uniform.rvs() 
    if Y < 0.4:
        T = expon.rvs(loc=0, scale=5) 
    else:
        T = uniform.rvs(loc=4, scale=2)
    return T

data = [getRandomTime() for _ in range(1000)]

print("Valoarea medie:", mean(data))
print("Deviatia standard:", std(data))

prob = [d < 5 for d in data]
print("Prob ca timpul < 5:", mean(prob))

print(
    "Prob teoretica:",
    expon.cdf(5, loc=0, scale=5) * 0.4
    + uniform.cdf(5, loc=4, scale=2) * 0.6
)
```

```python
from scipy.stats import uniform
from numpy import exp, mean
from scipy.integrate import quad
n = 1000000
U = uniform.rvs(loc=-1, scale=4, size=n)
g = exp(-U**2)
MC = 4 * mean(g)
print("Aprox Monte Carlo", MC)
exact, _ = quad(lambda x: exp(-x**2), -1, 3)
print("Valoare exacta", exact)
```
