## 1. Definirea problemei

### 1.1 Context și motivație

Arhitectura software modernă se bazează pe diagrame pentru comunicarea structurii sistemelor. Totuși, există o problemă fundamentală: **documentația se desincronizează rapid cu implementarea**.

**Cauze identificate:**
- Diagramele sunt create manual, consumând timp
- Nu există un mecanism automat de detectare a schimbărilor arhitecturale
- Echipele neglijează actualizarea documentației sub presiunea termenelor

### 1.2 Problema formalizată

**Input:** Descriere text a arhitecturii $D$ sau cod sursă Python $C$  
**Output:** Diagramă Mermaid $M$ care respectă sintaxa DSL și reflectă corect arhitectura  

Dorim să antrenăm/fine-tunăm un model $f: (D \cup C) \rightarrow M$ care maximizează:
- $P(\text{valid}(M)) > 0.9$ — diagrama este sintactic validă
- $\text{BLEU}(M, M^*) > 0.6$ — similaritate cu ground truth
- $F_1(\text{nodes}(M), \text{nodes}(M^*)) > 0.7$ — noduri corecte

### 1.3 Tipuri de diagrame țintă

| Tip | Utilizare | Prioritate |
|-----|-----------|------------|
| `flowchart` | Fluxuri de date, procese | ⭐⭐⭐ |
| `classDiagram` | Structuri OOP | ⭐⭐⭐ |
| `sequenceDiagram` | Interacțiuni în timp | ⭐⭐ |
| `C4Context` | Arhitecturi la nivel înalt | ⭐⭐ |

## 3. Analiza Exploratorie a Datelor (EDA)

### 3.1 Statistici generale ale dataset-ului

````python
# Calculăm statistici detaliate pentru fiecare diagramă
def extract_diagram_stats(row):
    code = row['mermaid_code']
    lines = [l.strip() for l in code.split('\n') if l.strip()]
    edge_count = sum(1 for l in lines if re.search(r'-->|---|->>|==>|--|\|', l))
    node_ids = set()
    for l in lines[1:]:
        matches = re.findall(r'\b([A-Za-z_]\w*)\s*(?:\[|\(|\{|-->|---)', l)
        node_ids.update(matches)
    return pd.Series({
        'num_lines': len(lines),
        'num_edges': edge_count,
        'num_nodes': len(node_ids),
        'code_length': len(code),
        'desc_length': len(row['description'].split()),
        'has_condition': 1 if '{' in code or 'alt' in code.lower() else 0,
    })

stats_df = df.join(df.apply(extract_diagram_stats, axis=1))

print('=== Statistici descriptive ===')
print(stats_df[['num_lines', 'num_edges', 'num_nodes', 'code_length', 'desc_length']]
      .describe().round(2).to_string())
````

````python
fig, axes = plt.subplots(2, 3, figsize=(15, 9))
fig.suptitle('EDA — Analiză dataset diagrame Mermaid', fontsize=14, fontweight='bold')

# 1. Distribuție tipuri
type_counts = df['diagram_type'].value_counts()
colors = sns.color_palette('husl', len(type_counts))
axes[0,0].bar(type_counts.index, type_counts.values, color=colors, edgecolor='white')
axes[0,0].set_title('Distribuție tipuri diagrame')
axes[0,0].set_ylabel('Număr exemple')
for i, v in enumerate(type_counts.values):
    axes[0,0].text(i, v + 0.05, str(v), ha='center', fontweight='bold')

# 2. Distribuție surse
src_counts = df['source'].value_counts()
axes[0,1].pie(src_counts.values, labels=src_counts.index,
              autopct='%1.0f%%', colors=sns.color_palette('pastel'))
axes[0,1].set_title('Surse date')

# 3. Lungime cod per tip
for dtype in stats_df['diagram_type'].unique():
    subset = stats_df[stats_df['diagram_type'] == dtype]['num_lines']
    axes[0,2].scatter([dtype] * len(subset), subset, alpha=0.7, s=80)
axes[0,2].set_title('Linii cod per tip diagramă')
axes[0,2].set_ylabel('Număr linii')

# 4. Număr noduri vs edge-uri
scatter = axes[1,0].scatter(stats_df['num_nodes'], stats_df['num_edges'],
                            c=stats_df['diagram_type'].astype('category').cat.codes,
                            cmap='tab10', s=100, alpha=0.8)
axes[1,0].set_xlabel('Număr noduri')
axes[1,0].set_ylabel('Număr edge-uri')
axes[1,0].set_title('Noduri vs Edge-uri')

# 5. Distribuție lungime descrieri
axes[1,1].hist(stats_df['desc_length'], bins=10, color='steelblue', edgecolor='white')
axes[1,1].set_title('Lungime descrieri (cuvinte)')
axes[1,1].set_xlabel('Număr cuvinte')
axes[1,1].axvline(stats_df['desc_length'].mean(), color='red', linestyle='--', label=f'Media: {stats_df["desc_length"].mean():.1f}')
axes[1,1].legend()

# 6. Complexitate (linii cod) per tip
stats_df.boxplot(column='num_lines', by='diagram_type', ax=axes[1,2])
axes[1,2].set_title('Complexitate per tip (linii cod)')
axes[1,2].set_xlabel('')
plt.sca(axes[1,2])
plt.xticks(rotation=30)

plt.tight_layout()
plt.savefig('diagram_ai/data/processed/eda_overview.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Grafic salvat: data/processed/eda_overview.png')
````

### 3.2 Analiza structurală a diagramelor

Analizăm pattern-urile structurale — ce tipuri de noduri și relații apar mai frecvent.

````python
# Analiza vocabularului DSL Mermaid
arrow_types = Counter()
node_shapes = Counter()

ARROW_PATTERNS = {
    '-->': 'solid arrow', '---': 'solid line', '->>': 'async arrow',
    '-->>': 'dashed arrow', '-.->': 'dotted arrow', '==>': 'thick arrow'
}
SHAPE_PATTERNS = {
    r'\[': 'rectangle', r'\(': 'rounded rect',
    r'\{': 'diamond/decision', r'\[\[': 'subroutine'
}

for _, row in df.iterrows():
    code = row['mermaid_code']
    for pattern, name in ARROW_PATTERNS.items():
        arrow_types[name] += code.count(pattern)
    for pattern, name in SHAPE_PATTERNS.items():
        node_shapes[name] += len(re.findall(pattern, code))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('Vocabular DSL Mermaid în dataset', fontweight='bold')

# Frecvența tipurilor de săgeți
arrows = dict(sorted(arrow_types.items(), key=lambda x: x[1], reverse=True))
bars = ax1.barh(list(arrows.keys()), list(arrows.values()), color=sns.color_palette('Blues_r', len(arrows)))
ax1.set_title('Frecvența tipurilor de relații')
ax1.set_xlabel('Număr apariții')
for bar, val in zip(bars, arrows.values()):
    ax1.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, str(val), va='center')

# Forme noduri
shapes = dict(sorted(node_shapes.items(), key=lambda x: x[1], reverse=True))
ax2.bar(list(shapes.keys()), list(shapes.values()), color=sns.color_palette('Greens_r', len(shapes)), edgecolor='white')
ax2.set_title('Frecvența formelor de noduri')
ax2.set_ylabel('Număr apariții')
plt.xticks(rotation=20)

plt.tight_layout()
plt.savefig("diagram_ai/data/processed/eda_dsl_vocab.png", dpi=150, bbox_inches='tight')
plt.show()
````

## 4. Analiza Codului Sursă Python (AST)

### 4.1 Demonstrarea analizei AST statice

Unul din aspectele distinctive ale proiectului este generarea diagramelor **din cod sursă** prin analiză statică AST.

````python
from diagram_ai.src.ast_analyzer.analyzer import PythonASTAnalyzer

# Cod Python de test — un sistem de e-commerce simplificat
SAMPLE_CODE = '''
from dataclasses import dataclass
from typing import Optional
import uuid

class Repository:
    """Base repository with CRUD operations."""
    def find_by_id(self, id: str):
        pass
    def save(self, entity):
        pass
    def delete(self, id: str):
        pass

class UserRepository(Repository):
    def find_by_email(self, email: str):
        return self.db.query(User).filter_by(email=email).first()
    def find_active(self):
        return self.db.query(User).filter_by(is_active=True).all()

class OrderRepository(Repository):
    def find_by_user(self, user_id: str):
        return self.db.query(Order).filter_by(user_id=user_id).all()
    def find_pending(self):
        return self.db.query(Order).filter_by(status="pending").all()

class AuthService:
    def __init__(self, user_repo: UserRepository):
        self.user_repo = user_repo
    def authenticate(self, email: str, password: str) -> Optional[str]:
        user = self.user_repo.find_by_email(email)
        if user and self.verify_password(password, user.hashed_password):
            return self.generate_jwt(user.id)
        return None
    def generate_jwt(self, user_id: str) -> str:
        return str(uuid.uuid4())
    def verify_password(self, plain: str, hashed: str) -> bool:
        return plain == hashed

class OrderService:
    def __init__(self, order_repo: OrderRepository):
        self.order_repo = order_repo
    def create_order(self, user_id: str, items: list) -> str:
        order_id = str(uuid.uuid4())
        self.order_repo.save({"id": order_id, "user_id": user_id, "items": items})
        return order_id
    def get_user_orders(self, user_id: str) -> list:
        return self.order_repo.find_by_user(user_id)
'''

analyzer = PythonASTAnalyzer()
module_info = analyzer.analyze_source(SAMPLE_CODE, name='ecommerce')

print('=== Rezultat analiză AST statică ===')
print(f'Modul: {module_info.name}')
print(f'Importuri: {module_info.imports}')
print(f'\nClase detectate ({len(module_info.classes)}):')
for cls in module_info.classes:
    bases_str = f' (extends: {", ".join(cls.bases)})' if cls.bases else ''
    print(f'  📦 {cls.name}{bases_str}')
    print(f'     Metode: {cls.methods}')
````

````python
# Generare automată diagrame din AST
class_diagram = analyzer.to_mermaid_class_diagram(module_info)
flow_diagram = analyzer.to_mermaid_flowchart(module_info)

print('=== Class Diagram generat din AST ===')
print('```mermaid')
print(class_diagram)
print('```')

print('\n=== Flowchart generat din AST ===')
print('```mermaid')
print(flow_diagram)
print('```')

print('\n💡 Copiați codul de mai sus la https://mermaid.live pentru a vedea diagrama rendered')
````

## 5. Analiza RAG Engine — Calitatea Retrieval-ului

### 5.1 Indexare și testare retrieval

````python
from diagram_ai.src.rag_engine.engine import RAGEngine, DiagramExample

# Inițializare RAG engine
os.makedirs('../data/chromadb', exist_ok=True)
rag = RAGEngine(db_path='../data/chromadb')

# Indexare dataset seed
examples = [
    DiagramExample(
        doc_id=item['id'],
        description=item['description'],
        diagram_type=item['diagram_type'],
        mermaid_code=item['mermaid_code'],
        source=item['source'],
        metadata={},
    )
    for item in SEED_DATASET
]

count = rag.index_examples(examples)
print(f'✅ Indexate {count} diagrame')
print(f'Statistici: {rag.get_stats()}')
````

````python
# Test retrieval — query-uri de test
TEST_QUERIES = [
    ("Authentication system with JWT and refresh tokens", "flowchart"),
    ("Class hierarchy for repository pattern", "classDiagram"),
    ("API request handling sequence", "sequenceDiagram"),
    ("Deployment pipeline with Docker", "flowchart"),
]

print('=== Test Retrieval RAG ===\n')
retrieval_results = []

for query, dtype in TEST_QUERIES:
    result = rag.retrieve(query=query, diagram_type=dtype, top_k=3)
    top_sim = 1 - result.distances[0] if result.distances else 0
    top_desc = result.examples[0].description if result.examples else 'N/A'
    
    retrieval_results.append({
        'query': query[:40] + '...',
        'type': dtype,
        'top_similarity': round(top_sim, 3),
        'top_result': top_desc[:40] + '...'
    })
    
    print(f'📍 Query: "{query[:50]}"')
    for i, (ex, dist) in enumerate(zip(result.examples, result.distances), 1):
        sim = 1 - dist
        marker = '🥇' if i == 1 else '  '
        print(f'   {marker} [{sim:.3f}] {ex.description[:60]}')
    print()

retrieval_df = pd.DataFrame(retrieval_results)
print('\n=== Rezumat retrieval ===')
print(retrieval_df.to_string(index=False))
````

````python
# Vizualizare heatmap similaritate
n = len(SEED_DATASET)
queries_for_heatmap = [item['description'] for item in SEED_DATASET[:8]]
docs_for_heatmap = queries_for_heatmap

# Matrix de similaritate
sim_matrix = np.zeros((len(queries_for_heatmap), len(docs_for_heatmap)))

for i, q in enumerate(queries_for_heatmap):
    result = rag.retrieve(query=q, top_k=len(docs_for_heatmap))
    for j, (ex, dist) in enumerate(zip(result.examples, result.distances)):
        # Găsim indexul documentului
        for k, doc in enumerate(docs_for_heatmap):
            if doc == ex.description:
                sim_matrix[i, k] = 1 - dist
                break

# Diagrama diagonal trebuie să fie 1.0 (self-similarity)
for i in range(min(len(queries_for_heatmap), len(docs_for_heatmap))):
    sim_matrix[i, i] = 1.0

plt.figure(figsize=(10, 8))
short_labels = [q[:30] + '...' for q in queries_for_heatmap]
sns.heatmap(sim_matrix, annot=True, fmt='.2f', cmap='RdYlGn',
            xticklabels=short_labels, yticklabels=short_labels,
            vmin=0, vmax=1, linewidths=0.5)
plt.title('Matrice similaritate cosine — Sentence-BERT embeddings', fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.savefig('diagram_ai\data\processed\eda_similarity_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Heatmap salvat: data/processed/eda_similarity_heatmap.png')
````

## 6. Calitatea Datelor și Validare Sintaxă

### 6.1 Validare sintaxă Mermaid pe întreg dataset-ul

````python
from diagram_ai.src.utils.validator import MermaidValidator

validator = MermaidValidator()
validation_results = []

for item in SEED_DATASET:
    result = validator.validate_detailed(item['mermaid_code'])
    validation_results.append({
        'id': item['id'],
        'type': item['diagram_type'],
        'valid': result.is_valid,
        'errors': len(result.errors),
        'warnings': len(result.warnings),
        'nodes': result.node_count,
        'edges': result.edge_count,
    })

val_df = pd.DataFrame(validation_results)
valid_pct = val_df['valid'].mean() * 100

print(f'=== Calitate dataset ===')
print(f'Diagrame valide sintactic: {val_df["valid"].sum()}/{len(val_df)} ({valid_pct:.1f}%)')
print(f'\nMedii per tip:')
print(val_df.groupby('type')[['valid', 'nodes', 'edges']].mean().round(2).to_string())

invalid = val_df[~val_df['valid']]
if len(invalid) > 0:
    print(f'\n⚠️  Diagrame invalide:\n{invalid["id"].tolist()}')
else:
    print('\n✅ Toate diagramele din dataset sunt valide sintactic!')
````
