```python
#3
from random import sample
from math import perm,comb
from itertools import permutations,combinations
```

```python
def aranjamente(word,k,numar_total=False,aleator=False):
    if numar_total:
        return perm(len(word),k)
    elif aleator:
        return sample(word,k)
    else:
        return list(permutations(word,k))
```

```python
aranjamente('word',2)
```

```python
aranjamente('word',2,numar_total=True)
```

```python
aranjamente('word',2,aleator=True)
```

```python
def combinari(word,k,numar_total=False,aleator=False):
    if numar_total:
        return comb(len(word),k)
    elif aleator:
        return sample(list(combinations(word,k)),1)
    else:
        return list(combinations(word,k))
```

```python
combinari('word',2)
```

```python
combinari('word',2,numar_total=True)
```

```python
combinari('word',2,aleator=True)
```
