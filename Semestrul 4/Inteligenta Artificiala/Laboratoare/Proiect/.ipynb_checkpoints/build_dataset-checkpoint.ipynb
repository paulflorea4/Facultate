## Cell 2 — Imports

```python
import ast
import json
import re
from pathlib import Path
from typing import Optional
from tqdm.notebook import tqdm
```

## Cell 3 — AST Extractor

```python
class UMLDataExtractor(ast.NodeVisitor):
    """
    Walks a Python AST and extracts classes, methods,
    fields, inheritance, and inter-class associations.
    """
    def __init__(self):
        self.metadata = {"classes": []}
        self._class_names: set = set()

    def visit_ClassDef(self, node):
        bases = [b.id for b in node.bases if isinstance(b, ast.Name)]
        methods, fields = [], []

        for item in node.body:
            if isinstance(item, ast.FunctionDef):
                access = "-" if item.name.startswith("_") else "+"
                params = [a.arg for a in item.args.args if a.arg != "self"]
                ret = ""
                if item.returns and isinstance(item.returns, ast.Name):
                    ret = f" : {item.returns.id}"
                methods.append(f"{access}{item.name}({', '.join(params)}){ret}")

            elif isinstance(item, ast.AnnAssign) and isinstance(item.target, ast.Name):
                ann = ""
                if isinstance(item.annotation, ast.Name):
                    ann = item.annotation.id
                elif isinstance(item.annotation, ast.Subscript):
                    try:
                        ann = ast.unparse(item.annotation)
                    except Exception:
                        ann = ""
                fields.append({"name": item.target.id, "type": ann})

        self.metadata["classes"].append({
            "name":     node.name,
            "inherits": bases,
            "methods":  methods,
            "fields":   fields,
        })
        self._class_names.add(node.name)
        self.generic_visit(node)

    def post_process(self):
        """Detect associations via type-annotated method parameters."""
        known = self._class_names
        for cls in self.metadata["classes"]:
            assoc = set()
            for m in cls["methods"]:
                for k in known:
                    if k in m and k != cls["name"]:
                        assoc.add(k)
            cls["associations"] = list(assoc)
        return self.metadata


def extract(source: str) -> Optional[dict]:
    """Parse source and return metadata dict, or None on failure."""
    try:
        tree = ast.parse(source)
    except SyntaxError:
        return None
    ex = UMLDataExtractor()
    ex.visit(tree)
    return ex.post_process()


# Quick sanity check
test_src = '''
class Animal:
    name: str
    def speak(self): pass

class Dog(Animal):
    breed: str
    def fetch(self): pass
    def _bark(self): pass
'''
print(json.dumps(extract(test_src), indent=2))
```

## Cell 4 — Diagram templating functions

```python
def to_mermaid(meta: dict) -> str:
    lines = ["classDiagram"]
    for cls in meta["classes"]:
        lines.append(f"    class {cls['name']} {{")
        for f in cls.get("fields", []):
            t = f" {f['type']}" if f["type"] else ""
            lines.append(f"        +{f['name']}{t}")
        for m in cls["methods"]:
            lines.append(f"        {m}")
        lines.append("    }")
    for cls in meta["classes"]:
        for base in cls["inherits"]:
            lines.append(f"    {base} <|-- {cls['name']}")
        for assoc in cls.get("associations", []):
            lines.append(f"    {cls['name']} --> {assoc}")
    return "\n".join(lines)


def to_plantuml(meta: dict) -> str:
    lines = ["@startuml", ""]
    for cls in meta["classes"]:
        kw = "abstract class" if cls["name"].startswith(("Abstract", "Base")) else "class"
        lines.append(f"{kw} {cls['name']} {{")
        for f in cls.get("fields", []):
            t = f" : {f['type']}" if f["type"] else ""
            lines.append(f"    +{f['name']}{t}")
        for m in cls["methods"]:
            lines.append(f"    {m}")
        lines.append("}")
        lines.append("")
    for cls in meta["classes"]:
        for base in cls["inherits"]:
            lines.append(f"{base} <|-- {cls['name']}")
        for assoc in cls.get("associations", []):
            lines.append(f"{cls['name']} --> {assoc}")
    lines += ["", "@enduml"]
    return "\n".join(lines)


def to_description(meta: dict) -> str:
    """Generate a natural language description of the class structure."""
    parts = []
    for cls in meta["classes"]:
        desc = f"The `{cls['name']}` class"
        if cls["inherits"]:
            desc += f" extends {', '.join(cls['inherits'])}"
        pub  = [m.lstrip("+").split("(")[0] for m in cls["methods"] if m.startswith("+")]
        priv = [m.lstrip("-").split("(")[0] for m in cls["methods"] if m.startswith("-")]
        if pub:
            desc += f" and exposes public methods: {', '.join(pub)}"
        if priv:
            desc += f". Private methods: {', '.join(priv)}"
        if cls.get("fields"):
            desc += f". Attributes: {', '.join(f['name'] for f in cls['fields'])}"
        if cls.get("associations"):
            desc += f". Collaborates with: {', '.join(cls['associations'])}"
        parts.append(desc + ".")
    return " ".join(parts)


# Quick check
meta = extract(test_src)
print("--- Mermaid ---")
print(to_mermaid(meta))
print("\n--- PlantUML ---")
print(to_plantuml(meta))
print("\n--- Description ---")
print(to_description(meta))
```

## Cell 5 — Quality filter

```python
def is_useful(meta: dict) -> bool:
    """
    Returns True only if the file will produce a meaningful diagram.

    Rules:
      - At least 2 classes (a single class produces a trivial diagram)
      - At least one relationship (inheritance or association)
      - Every class has at least 1 method
      - No class name is a single letter (likely auto-generated or test stubs)
    """
    classes = meta.get("classes", [])

    if len(classes) < 2:
        return False

    has_relationship = any(
        cls["inherits"] or cls.get("associations")
        for cls in classes
    )
    if not has_relationship:
        return False

    if not all(len(cls["methods"]) >= 1 for cls in classes):
        return False

    if any(len(cls["name"]) <= 1 for cls in classes):
        return False

    return True


# Test the filter
print("Is useful:", is_useful(meta))   # should be True for our dog/animal example
```

## Cell 6 — Configuration
**Edit these values before running the pipeline.**

```python
# ── How many raw samples to pull from The Stack ──────────────────────────────
# Not all will pass the quality filter, so pull more than you need.
# ~30% of files typically pass → to get 1000 pairs, pull ~3500 samples.
NUM_SAMPLES = 3500

# ── Output file paths ─────────────────────────────────────────────────────────
OUT_FULL     = "data/dataset_full.jsonl"       # one record per file, all formats
OUT_MERMAID  = "data/dataset_instruct_mermaid.jsonl"   # instruction-tuning, Mermaid output
OUT_PLANTUML = "data/dataset_instruct_plantuml.jsonl"  # instruction-tuning, PlantUML output
OUT_NL       = "data/dataset_instruct_nl.jsonl"        # description → mermaid pairs

# ── Stats tracking ────────────────────────────────────────────────────────────
stats = {
    "pulled":        0,
    "parse_error":   0,
    "filtered_out":  0,
    "saved":         0,
}

print("Config ready.")
print(f"  Will pull up to {NUM_SAMPLES} samples from The Stack")
print(f"  Full dataset  → {OUT_FULL}")
print(f"  Mermaid pairs → {OUT_MERMAID}")
print(f"  PlantUML pairs→ {OUT_PLANTUML}")
print(f"  NL pairs      → {OUT_NL}")
```

## Cell 7 — Load The Stack (streaming, no full download needed)

```python
import os
from huggingface_hub import login

login()
print("Logged in.")
```

```python
from datasets import load_dataset

ds = load_dataset(
    "bigcode/the-stack",
    data_dir="data/python",
    split="train",
    streaming=True,
    token=HF_TOKEN,        # passes auth
)

print("Dataset loaded in streaming mode.")
print("Sample keys:", list(next(iter(ds)).keys()))
```

## Cell 8 — Run the pipeline

```python
f_full     = open(OUT_FULL,     "w", encoding="utf-8")
f_mermaid  = open(OUT_MERMAID,  "w", encoding="utf-8")
f_plantuml = open(OUT_PLANTUML, "w", encoding="utf-8")
f_nl       = open(OUT_NL,       "w", encoding="utf-8")

try:
    for sample in tqdm(ds, total=NUM_SAMPLES, desc="Processing"):
        if stats["pulled"] >= NUM_SAMPLES:
            break

        stats["pulled"] += 1
        source = sample.get("content", "")
        repo   = sample.get("max_stars_repo_name", "unknown")

        # Step 1: extract
        meta = extract(source)
        if meta is None:
            stats["parse_error"] += 1
            continue

        # Step 2: filter
        if not is_useful(meta):
            stats["filtered_out"] += 1
            continue

        # Step 3: generate all formats
        mermaid     = to_mermaid(meta)
        plantuml    = to_plantuml(meta)
        description = to_description(meta)

        # Step 4: write full record (all formats in one line)
        full_record = {
            "source_code": source,
            "description": description,
            "mermaid":     mermaid,
            "plantuml":    plantuml,
            "metadata":    meta,
            "repo":        repo,
        }
        f_full.write(json.dumps(full_record, ensure_ascii=False) + "\n")

        # Step 5: write instruction-tuning records
        # — code → mermaid
        f_mermaid.write(json.dumps({
            "instruction": "Generate a Mermaid class diagram for the following Python code.",
            "input":       source,
            "output":      mermaid,
        }, ensure_ascii=False) + "\n")

        # — code → plantuml
        f_plantuml.write(json.dumps({
            "instruction": "Generate a PlantUML class diagram for the following Python code.",
            "input":       source,
            "output":      plantuml,
        }, ensure_ascii=False) + "\n")

        # — natural language description → mermaid
        f_nl.write(json.dumps({
            "instruction": "Generate a Mermaid class diagram from this architecture description.",
            "input":       description,
            "output":      mermaid,
        }, ensure_ascii=False) + "\n")

        stats["saved"] += 1

finally:
    f_full.close()
    f_mermaid.close()
    f_plantuml.close()
    f_nl.close()

print("\n✓ Pipeline finished")
print(f"  Pulled:       {stats['pulled']}")
print(f"  Parse errors: {stats['parse_error']}")
print(f"  Filtered out: {stats['filtered_out']}")
print(f"  Saved:        {stats['saved']} training pairs")
```

## Cell 9 — Inspect a few records

```python
# Read back the first 3 records from the full dataset and pretty-print them
records = []
with open(OUT_FULL, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i >= 3:
            break
        records.append(json.loads(line))

for i, r in enumerate(records, 1):
    print(f"{'='*60}")
    print(f"RECORD {i}  |  repo: {r['repo']}")
    print(f"{'='*60}")
    print("[Description]")
    print(r["description"])
    print("\n[Mermaid]")
    print(r["mermaid"])
    print("\n[Classes extracted]", len(r["metadata"]["classes"]))
    print()
```

## Cell 10 — Dataset statistics

```python
from collections import Counter

class_counts      = []
method_counts     = []
inheritance_counts = []
has_fields        = 0

with open(OUT_FULL, "r", encoding="utf-8") as f:
    for line in f:
        r = json.loads(line)
        classes = r["metadata"]["classes"]
        class_counts.append(len(classes))
        for cls in classes:
            method_counts.append(len(cls["methods"]))
            inheritance_counts.append(len(cls["inherits"]))
            if cls.get("fields"):
                has_fields += 1

total = len(class_counts)
print(f"Total training pairs : {total}")
print(f"Total classes        : {sum(class_counts)}")
print(f"Avg classes/file     : {sum(class_counts)/total:.1f}")
print(f"Avg methods/class    : {sum(method_counts)/len(method_counts):.1f}")
print(f"Classes with fields  : {has_fields}")
print(f"Classes with inherit : {sum(1 for x in inheritance_counts if x > 0)}")

# Distribution of class counts per file
dist = Counter(class_counts)
print("\nClasses per file distribution:")
for k in sorted(dist)[:10]:
    print(f"  {k} classes: {dist[k]} files")
```
