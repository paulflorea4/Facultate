```python
from scipy.stats import bernoulli,binom,geom,hypergeom,randint,uniform,norm
```

```python
uniform.rvs(loc=1,scale=9,size=10)
```

```python
uniform.pdf(x=0.4,loc=0,scale=1)
```

```python
uniform.cdf(x=0.4,loc=0,scale=1)
```

```python
norm.rvs(loc=1.65,scale=0.10,size=100)
```

```python
norm.pdf(x=1.7,loc=1.65,scale=0.10)
```

```python
bernoulli.rvs(p=2/3,size=100)
```

```python
bernoulli.pmf(k=0,p=1/3)
```

```python
bernoulli.cdf(k=0,p=1/3)
```

```python
binom.rvs(n=10,p=1/2,size=100)
```

```python
binom.pmf(k=9,n=10,p=0.5)
```

```python
geom.rvs(p=1/2,size=100)
```

```python
geom.pmf(3,0.5)
```

```python
geom.cdf(3,0.5)
```

```python
hypergeom.rvs(M=20,n=5,N=4,size=100)
```

```python
randint.rvs(1,10,size=10)
```

```python
randint.cdf(3,1,11)
```
