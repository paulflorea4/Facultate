```python
import ast
import json
import os
import time
import hashlib
import warnings
from pathlib import Path
from typing import Optional
from collections import Counter, defaultdict
from tqdm.notebook import tqdm

# Suprimam SyntaxWarning-urile din fisierele the-stack
# (regex-uri cu escape sequences invalide din cod Python 2/vechi)
warnings.filterwarnings("ignore", category=SyntaxWarning)

print("✓ Imports OK")
```

## 1. Extractor AST — Versiunea Completă

Față de v1, această versiune extrage și:
- **Decoratori** pe clase și metode (`@property`, `@staticmethod`, `@abstractmethod`)
- **Docstrings** ale claselor (utile pentru generarea descrierilor NL)
- **Asocieri compuse** (detectate și din type hints pe câmpuri, nu doar parametri)
- **Hash deduplication** — evităm perechile duplicate dacă același fișier apare în mai multe repo-uri

```python
class UMLDataExtractorV2(ast.NodeVisitor):
    """
    Extractor AST complet v2.
    Extrage clase, metode, câmpuri, moștenire, asocieri,
    decoratori și docstrings din cod Python.
    """
    def __init__(self):
        self.metadata = {"classes": []}
        self._class_names: set = set()

    def _get_decorators(self, node) -> list:
        decorators = []
        for d in node.decorator_list:
            if isinstance(d, ast.Name):
                decorators.append(d.id)
            elif isinstance(d, ast.Attribute):
                decorators.append(d.attr)
        return decorators

    def _get_docstring(self, node) -> str:
        if (node.body and isinstance(node.body[0], ast.Expr)
                and isinstance(node.body[0].value, ast.Constant)
                and isinstance(node.body[0].value.value, str)):
            return node.body[0].value.value.strip().split("\n")[0][:120]
        return ""

    def _resolve_annotation(self, annotation) -> str:
        """Converteste un nod de adnotare in string."""
        if annotation is None:
            return ""
        try:
            return ast.unparse(annotation)
        except Exception:
            return ""

    def visit_ClassDef(self, node):
        bases = []
        for base in node.bases:
            if isinstance(base, ast.Name):
                bases.append(base.id)
            elif isinstance(base, ast.Attribute):
                bases.append(base.attr)

        decorators = self._get_decorators(node)
        docstring  = self._get_docstring(node)
        methods, fields = [], []
        field_types = set()

        for item in node.body:
            if isinstance(item, ast.FunctionDef):
                access = "-" if item.name.startswith("_") and not item.name.startswith("__") else "+"
                params = []
                for a in item.args.args:
                    if a.arg == "self":
                        continue
                    ann = self._resolve_annotation(a.annotation)
                    params.append(f"{a.arg}: {ann}" if ann else a.arg)
                ret = ""
                if item.returns:
                    ret = f" : {self._resolve_annotation(item.returns)}"
                method_decorators = self._get_decorators(item)
                stereotype = ""
                if "abstractmethod" in method_decorators:
                    stereotype = "<<abstract>> "
                elif "staticmethod" in method_decorators:
                    stereotype = "<<static>> "
                elif "classmethod" in method_decorators:
                    stereotype = "<<class>> "
                methods.append(f"{access}{stereotype}{item.name}({', '.join(params)}){ret}")

            elif isinstance(item, ast.AnnAssign) and isinstance(item.target, ast.Name):
                ann = self._resolve_annotation(item.annotation)
                fields.append({"name": item.target.id, "type": ann})
                # Colectam tipurile pentru detectia asocierilor prin campuri
                if ann:
                    field_types.add(ann.split("[")[0].strip())

        self.metadata["classes"].append({
            "name":        node.name,
            "inherits":    bases,
            "methods":     methods,
            "fields":      fields,
            "decorators":  decorators,
            "docstring":   docstring,
            "_field_types": list(field_types),
        })
        self._class_names.add(node.name)
        self.generic_visit(node)

    def post_process(self):
        """Detecteaza asocieri prin parametri SI prin campuri tipizate."""
        known = self._class_names
        for cls in self.metadata["classes"]:
            assoc = set()
            # Din parametrii metodelor
            for m in cls["methods"]:
                for k in known:
                    if k in m and k != cls["name"]:
                        assoc.add(k)
            # Din campurile tipizate
            for ft in cls.get("_field_types", []):
                if ft in known and ft != cls["name"]:
                    assoc.add(ft)
            cls["associations"] = list(assoc)
            del cls["_field_types"]
        return self.metadata


def extract_v2(source: str) -> Optional[dict]:
    try:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", SyntaxWarning)
            tree = ast.parse(source)
    except SyntaxError:
        return None
    try:
        ex = UMLDataExtractorV2()
        ex.visit(tree)
        return ex.post_process()
    except RecursionError:
        return None


# Sanity check
test = '''
class Repository:
    """Generic data access layer."""
    db: Database
    def find(self, id: int) -> Model: pass
    def save(self, obj: Model): pass
    def _validate(self): pass

class Database:
    host: str
    port: int
    def connect(self): pass
    def execute(self, query: str) -> list: pass

class Model:
    id: int
    created_at: str
    def validate(self) -> bool: pass
    def to_dict(self) -> dict: pass
'''
print(json.dumps(extract_v2(test), indent=2, ensure_ascii=False))
```

## 2. Template-uri DSL — Versiunea Completă

Față de v1, adăugăm:
- Stereotipuri pe clase (`<<abstract>>`, `<<interface>>`)
- Relații de compoziție vs. asociere simplă
- Notă cu docstring în diagramă (opțional)
- Format Structurizr DSL (bonus)

```python
def to_mermaid_v2(meta: dict) -> str:
    lines = ["classDiagram"]
    for cls in meta["classes"]:
        # Stereotype
        name = cls["name"]
        if "abstractmethod" in str(cls.get("decorators", "")) or name.startswith(("Abstract", "Base", "Interface")):
            lines.append(f"    class {name} {{")
            lines.append(f"        <<abstract>>")
        else:
            lines.append(f"    class {name} {{")
        for f in cls.get("fields", []):
            t = f" {f['type']}" if f["type"] else ""
            lines.append(f"        +{f['name']}{t}")
        for m in cls["methods"]:
            lines.append(f"        {m}")
        lines.append("    }")
    # Relatii
    for cls in meta["classes"]:
        for base in cls["inherits"]:
            lines.append(f"    {base} <|-- {cls['name']}")
        for assoc in cls.get("associations", []):
            # Daca clasa are un camp de tipul asociat → compozitie
            field_names = [f["type"] for f in cls.get("fields", [])]
            arrow = "*--" if assoc in field_names else "-->"
            lines.append(f"    {cls['name']} {arrow} {assoc}")
    return "\n".join(lines)


def to_plantuml_v2(meta: dict) -> str:
    lines = ["@startuml", "skinparam classAttributeIconSize 0", ""]
    for cls in meta["classes"]:
        name = cls["name"]
        kw = "abstract class" if name.startswith(("Abstract", "Base")) else "class"
        lines.append(f"{kw} {name} {{")
        if cls.get("docstring"):
            lines.append(f"  .. {cls['docstring'][:60]} ..")
        for f in cls.get("fields", []):
            t = f" : {f['type']}" if f["type"] else ""
            lines.append(f"  +{f['name']}{t}")
        for m in cls["methods"]:
            lines.append(f"  {m}")
        lines.append("}\n")
    for cls in meta["classes"]:
        for base in cls["inherits"]:
            lines.append(f"{base} <|-- {cls['name']}")
        for assoc in cls.get("associations", []):
            lines.append(f"{cls['name']} --> {assoc}")
    lines += ["", "@enduml"]
    return "\n".join(lines)


def to_description_v2(meta: dict) -> str:
    parts = []
    for cls in meta["classes"]:
        desc = f"The `{cls['name']}` class"
        if cls.get("docstring"):
            desc += f" ({cls['docstring']})"
        if cls["inherits"]:
            desc += f" extends {', '.join(cls['inherits'])}"
        pub  = [m.split("(")[0].lstrip("+") for m in cls["methods"] if m.startswith("+")]
        priv = [m.split("(")[0].lstrip("-") for m in cls["methods"] if m.startswith("-")]
        if pub:  desc += f" and exposes: {', '.join(pub)}"
        if priv: desc += f". Private: {', '.join(priv)}"
        if cls.get("fields"):
            desc += f". Fields: {', '.join(f['name'] + (':' + f['type'] if f['type'] else '') for f in cls['fields'])}"
        if cls.get("associations"):
            desc += f". Uses: {', '.join(cls['associations'])}"
        parts.append(desc + ".")
    return " ".join(parts)


# Test
meta_test = extract_v2(test)
print("--- Mermaid v2 ---")
print(to_mermaid_v2(meta_test))
print("\n--- PlantUML v2 ---")
print(to_plantuml_v2(meta_test))
```

## 3. Filtru de Calitate Îmbunătățit

Față de v1, adăugăm criterii suplimentare pentru a îmbunătăți calitatea perechilor:
- Excludem fișierele de test (`Test`, `Mock`, `Fake` în numele claselor)
- Minimum de complexitate (suma metodelor + câmpurilor ≥ 4)
- Deduplicare prin hash MD5 al codului sursă (evităm repo-uri care copiază cod)

```python
TEST_PATTERNS = {"Test", "Mock", "Fake", "Stub", "Dummy", "Fixture", "Spy"}
seen_hashes: set = set()   # pentru deduplicare globala

def is_useful_v2(meta: dict, source: str) -> tuple[bool, str]:
    """
    Returneaza (True, '') daca fisierul trece toate criteriile,
    altfel (False, motiv) pentru statistici de debugging.
    """
    classes = meta.get("classes", [])

    # Criteriu 1: minim 2 clase
    if len(classes) < 2:
        return False, "too_few_classes"

    # Criteriu 2: cel putin o relatie
    if not any(cls["inherits"] or cls.get("associations") for cls in classes):
        return False, "no_relationships"

    # Criteriu 3: fiecare clasa are minim 1 metoda
    if not all(len(cls["methods"]) >= 1 for cls in classes):
        return False, "empty_class"

    # Criteriu 4: niciun nume de clasa de 1 caracter
    if any(len(cls["name"]) <= 1 for cls in classes):
        return False, "single_char_name"

    # Criteriu 5 (NOU): excludem fisierele de test
    for cls in classes:
        for pattern in TEST_PATTERNS:
            if pattern in cls["name"]:
                return False, "test_file"

    # Criteriu 6 (NOU): complexitate minima — suma metode + campuri >= 4 per clasa
    for cls in classes:
        complexity = len(cls["methods"]) + len(cls.get("fields", []))
        if complexity < 2:
            return False, "low_complexity"

    # Criteriu 7 (NOU): deduplicare prin hash
    src_hash = hashlib.md5(source.encode()).hexdigest()
    if src_hash in seen_hashes:
        return False, "duplicate"
    seen_hashes.add(src_hash)

    return True, ""


# Test filtru
ok, reason = is_useful_v2(meta_test, test)
print(f"Repository/Database/Model example → {'✓ PASS' if ok else '✗ FAIL: ' + reason}")

# Test excludere test files
test_code = '''
class TestUser:
    def test_create(self): pass
class MockDatabase:
    def execute(self): pass
'''
meta_test2 = extract_v2(test_code)
ok2, reason2 = is_useful_v2(meta_test2, test_code)
print(f"TestUser/MockDatabase example    → {'✓ PASS' if ok2 else '✗ FAIL: ' + reason2}")
```

## 4. Configurare Pipeline

> ⚠️ **Editați valorile din celula de mai jos înainte de a rula pipeline-ul.**

Calculul: pentru ~5.000 perechi cu filtrul v2 (estimăm ~9% rată de trecere) avem nevoie de ~55.000 mostre brute.

```python
# ── EDITAȚI ACESTE VALORI ─────────────────────────────────────────────────────
HF_TOKEN    = "hf_YOUR_TOKEN_HERE"   # token HuggingFace cu acces la bigcode/the-stack
NUM_SAMPLES = 100_000                  # mostre brute; ~100k → ~5k perechi finale
TARGET_PAIRS = 5_000                  # oprim pipeline-ul dupa ce atingem tinta
CHECKPOINT_EVERY = 500                # salvam un checkpoint la fiecare N perechi

# ── Output paths ──────────────────────────────────────────────────────────────
Path("data").mkdir(exist_ok=True)
Path("data/checkpoints").mkdir(exist_ok=True)

OUT_FULL     = "data/dataset_full_v2.jsonl"
OUT_MERMAID  = "data/dataset_instruct_mermaid_v2.jsonl"
OUT_PLANTUML = "data/dataset_instruct_plantuml_v2.jsonl"
OUT_NL       = "data/dataset_instruct_nl_v2.jsonl"
STATS_FILE   = "data/pipeline_stats.json"

stats = {
    "pulled": 0, "parse_error": 0,
    "filter_reasons": defaultdict(int),
    "saved": 0, "start_time": time.time()
}

print("✓ Configurare OK")
print(f"  Mostre brute    : {NUM_SAMPLES:,}")
print(f"  Țintă perechi   : {TARGET_PAIRS:,}")
print(f"  Checkpoint ogni : {CHECKPOINT_EVERY} perechi")
print(f"  Output          : {OUT_FULL}")
```

## 5. Resume din Checkpoint

Dacă pipeline-ul s-a întrerupt, celula de mai jos detectează ultimul checkpoint și reia de acolo,
fără a reanaliza fișierele deja procesate.

```python
def load_checkpoint() -> tuple[int, int]:
    """
    Citeste ultimul checkpoint si returneaza (samples_skipped, pairs_saved).
    Reinitializeaza seen_hashes din datele existente.
    """
    checkpoints = sorted(Path("data/checkpoints").glob("ckpt_*.json"))
    if not checkpoints:
        print("Nu exista checkpoint — pornim de la zero.")
        return 0, 0

    latest = checkpoints[-1]
    with open(latest) as f:
        ckpt = json.load(f)

    # Reincarcam hashurile vazute
    if Path(OUT_FULL).exists():
        with open(OUT_FULL, encoding="utf-8") as f:
            for line in f:
                r = json.loads(line)
                h = hashlib.md5(r["source_code"].encode()).hexdigest()
                seen_hashes.add(h)

    print(f"✓ Checkpoint gasit: {latest.name}")
    print(f"  Salvate pana acum : {ckpt['saved']} perechi")
    print(f"  Samples procesate : {ckpt['pulled']}")
    print(f"  Hashes incarcate  : {len(seen_hashes)}")
    return ckpt["pulled"], ckpt["saved"]


skip_samples, already_saved = load_checkpoint()
stats["saved"] = already_saved
```

## 6. Autentificare HuggingFace și Încărcare Dataset

```python
import os
from huggingface_hub import login
from datasets import load_dataset

login()
print("✓ Autentificat pe HuggingFace")

ds = load_dataset(
    "bigcode/the-stack",
    data_dir="data/python",
    split="train",
    streaming=True,
)
print("✓ Dataset 'bigcode/the-stack' (Python) încărcat în mod streaming")
print("  Chei disponibile:", list(next(iter(ds)).keys()))
```

## 7. Pipeline Principal

Rulează pipeline-ul complet cu:
- Progresbar detaliat
- Salvare checkpoint automat la fiecare 500 perechi
- Statistici live pe categorii de filtrare
- Oprire automată la atingerea TARGET_PAIRS

```python
def save_checkpoint(stats: dict):
    ckpt_path = f"data/checkpoints/ckpt_{stats['saved']:05d}.json"
    with open(ckpt_path, "w") as f:
        json.dump({
            "pulled": stats["pulled"],
            "saved":  stats["saved"],
            "filter_reasons": dict(stats["filter_reasons"]),
            "elapsed_min": round((time.time() - stats["start_time"]) / 60, 1),
        }, f, indent=2)
    print(f"  💾 Checkpoint salvat: {ckpt_path}")


# Deschidem fisierele in append mode (pentru resume)
mode = "a" if already_saved > 0 else "w"
f_full     = open(OUT_FULL,     mode, encoding="utf-8")
f_mermaid  = open(OUT_MERMAID,  mode, encoding="utf-8")
f_plantuml = open(OUT_PLANTUML, mode, encoding="utf-8")
f_nl       = open(OUT_NL,       mode, encoding="utf-8")

pbar = tqdm(total=TARGET_PAIRS, initial=already_saved,
            desc="Perechi salvate", unit="pair")
sample_pbar = tqdm(total=NUM_SAMPLES, initial=skip_samples,
                   desc="Mostre analizate", unit="file", leave=False)

try:
    ds_iter = iter(ds)

    # Skip samples deja procesate (la resume)
    if skip_samples > 0:
        print(f"⏩ Sarind {skip_samples:,} samples deja procesate...")
        for _ in range(skip_samples):
            next(ds_iter)

    for sample in ds_iter:
        if stats["pulled"] + skip_samples >= NUM_SAMPLES:
            print("\n✓ Limita NUM_SAMPLES atinsa.")
            break
        if stats["saved"] >= TARGET_PAIRS:
            print(f"\n🎯 Tinta de {TARGET_PAIRS:,} perechi atinsa!")
            break

        stats["pulled"] += 1
        sample_pbar.update(1)

        source = sample.get("content", "")
        repo   = sample.get("max_stars_repo_name", "unknown")

        # Extragere AST
        meta = extract_v2(source)
        if meta is None:
            stats["parse_error"] += 1
            stats["filter_reasons"]["parse_error"] += 1
            continue

        # Filtru calitate
        ok, reason = is_useful_v2(meta, source)
        if not ok:
            stats["filter_reasons"][reason] += 1
            continue

        # Generare formate
        mermaid     = to_mermaid_v2(meta)
        plantuml    = to_plantuml_v2(meta)
        description = to_description_v2(meta)

        # Record complet
        full_record = {
            "source_code": source,
            "description": description,
            "mermaid":     mermaid,
            "plantuml":    plantuml,
            "metadata":    meta,
            "repo":        repo,
            "pair_id":     stats["saved"] + 1,
        }
        f_full.write(json.dumps(full_record, ensure_ascii=False) + "\n")

        f_mermaid.write(json.dumps({
            "instruction": "Generate a Mermaid class diagram for the following Python code.",
            "input":  source,
            "output": mermaid,
        }, ensure_ascii=False) + "\n")

        f_plantuml.write(json.dumps({
            "instruction": "Generate a PlantUML class diagram for the following Python code.",
            "input":  source,
            "output": plantuml,
        }, ensure_ascii=False) + "\n")

        f_nl.write(json.dumps({
            "instruction": "Generate a Mermaid class diagram from this architecture description.",
            "input":  description,
            "output": mermaid,
        }, ensure_ascii=False) + "\n")

        stats["saved"] += 1
        pbar.update(1)

        # Checkpoint automat
        if stats["saved"] % CHECKPOINT_EVERY == 0:
            for fh in [f_full, f_mermaid, f_plantuml, f_nl]:
                fh.flush()
            save_checkpoint(stats)

finally:
    for fh in [f_full, f_mermaid, f_plantuml, f_nl]:
        fh.close()
    pbar.close()
    sample_pbar.close()

elapsed = (time.time() - stats["start_time"]) / 60
print(f"\n{'='*50}")
print(f"✓ Pipeline finalizat în {elapsed:.1f} minute")
print(f"  Analizate   : {stats['pulled']:,}")
print(f"  Salvate     : {stats['saved']:,} perechi")
print(f"  Rată filtrare: {stats['saved']/max(stats['pulled'],1)*100:.1f}%")
print(f"\nMotive filtrare:")
for reason, count in sorted(stats['filter_reasons'].items(), key=lambda x: -x[1]):
    print(f"  {reason:<25}: {count:,}")

# Salvam statistici finale (doar campuri serializabile)
with open(STATS_FILE, "w") as f:
    json.dump({
        "pulled":         stats["pulled"],
        "saved":          stats["saved"],
        "parse_error":    stats["parse_error"],
        "filter_reasons": dict(stats["filter_reasons"]),
        "elapsed_min":    round(elapsed, 1),
    }, f, indent=2)
print(f"\n✓ Statistici salvate în {STATS_FILE}")
```

## 8. Verificare și Statistici Finale

```python
# ── Dimensiuni fișiere output ─────────────────────────────────────────────────
print("Fișiere generate:\n")
print(f"{'Fișier':<45} {'Linii':>7}  {'Mărime':>10}")
print("-" * 65)
for path in [OUT_FULL, OUT_MERMAID, OUT_PLANTUML, OUT_NL]:
    p = Path(path)
    if p.exists():
        lines = sum(1 for _ in open(p, encoding="utf-8"))
        size_mb = p.stat().st_size / 1024 / 1024
        print(f"  {path:<43} {lines:>7}  {size_mb:>8.1f} MB")
    else:
        print(f"  {path:<43}  (negăsit)")
```

```python
# ── EDA pe dataset-ul final ───────────────────────────────────────────────────
from collections import Counter

class_counts, method_counts, field_counts = [], [], []
rel_types = Counter()
top_repos = Counter()

with open(OUT_FULL, "r", encoding="utf-8") as f:
    for line in f:
        r = json.loads(line)
        classes = r["metadata"]["classes"]
        class_counts.append(len(classes))
        top_repos[r["repo"]] += 1
        has_inh = has_assoc = False
        for cls in classes:
            method_counts.append(len(cls["methods"]))
            field_counts.append(len(cls.get("fields", [])))
            if cls["inherits"]: has_inh = True
            if cls.get("associations"): has_assoc = True
        if has_inh and has_assoc: rel_types["inheritance + association"] += 1
        elif has_inh:             rel_types["inheritance only"] += 1
        elif has_assoc:           rel_types["association only"] += 1

total = len(class_counts)
print(f"Total perechi       : {total:,}")
print(f"Total clase         : {sum(class_counts):,}")
print(f"Medie clase/fișier  : {sum(class_counts)/total:.2f}")
print(f"Medie metode/clasă  : {sum(method_counts)/len(method_counts):.2f}")
print(f"Medie câmpuri/clasă : {sum(field_counts)/len(field_counts):.2f}")
print(f"\nDistribuție tipuri relații:")
for k, v in rel_types.most_common():
    pct = v / total * 100
    bar = "█" * int(pct / 2)
    print(f"  {k:<30} {bar:<50} {v:>5} ({pct:.1f}%)")
print(f"\nTop 5 repo-uri:")
for repo, cnt in top_repos.most_common(5):
    print(f"  {repo:<50} {cnt} fișiere")

print(f"\nDistribuție clase per fișier:")
dist = Counter(class_counts)
max_v = max(dist.values())
for k in sorted(dist)[:12]:
    bar = "█" * int(dist[k] / max_v * 35)
    print(f"  {k:>2} clase  {bar:<35} {dist[k]:>5}")
```
