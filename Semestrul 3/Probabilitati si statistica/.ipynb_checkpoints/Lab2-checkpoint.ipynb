```python
#1.a
from random import randint
import random

def same_birthday_simulations(n,nrSimulari):
    count=0
    for _ in range (nrSimulari):
        birthdays=[]
        for _ in range(n):
            day=random.randint(1,365)
            birthdays.append(day)

        bdaySet=set(birthdays)
        if len(birthdays)!=len(bdaySet):
            count+=1
            
    return count/nrSimulari
```

```python
same_birthday_simulations(23,10000)
```

```python
#1.b
def same_birthday(n):
      probability=1;
      for i in range(n):
          probability*=(365-i+1)/365

      return 1-probability
```

```python
same_birthday(15)
```

```python
#1.c
from matplotlib.pyplot import plot, grid, title, show

title('Plot test')
xs = range(2,51)
ys = [same_birthday_simulations(n,1000) for n in xs]
plot(xs,ys,'r*')
grid()
show()
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist
```

```python
#2.a

def puncte_patrat(N,cerinta):
    axis('square')
    axis((0,1,0,1))
    count=0
    A,B,C,D,E=[0,0],[1,0],[1,1],[0,1],[0.5,0.5]

    match cerinta:
        case 1:
            for _ in range(N):
                F=[random(),random()]
                if dist(F,E)<0.5:
                    count+=1
                    plot(F[0],F[1],'ro')
                else:
                    plot(F[0],F[1],'bo')
            return count/N
        case 2:
            for _ in range(N):
                F=[random(),random()]
                if dist(F,E)<min([dist(F,A),dist(F,B),dist(F,C),dist(F,D)]):
                    count+=1
                    plot(F[0],F[1],'ro')
                else:
                    plot(F[0],F[1],'bo')
            return count/N
        case 3:
            for _ in range(N):
                F=[random(),random()]
                nr_tri_obtuze=(dist(A,F)**2+dist(B,F)**2<1) + (dist(B,F)**2+dist(C,F)**2<1) + (dist(D,F)**2+dist(C,F)**2<1) + (dist(D,F)**2+dist(A,F)**2<1)
                if nr_tri_obtuze == 2:
                    count+=1
                    plot(F[0],F[1],'ro')
                else:
                    plot(F[0],F[1],'bo')
            return count/N
            
    
```

```python
#2.b
puncte_patrat(1000,1)
```

```python
puncte_patrat(1000,2)
```

```python
puncte_patrat(1000,3)
```
