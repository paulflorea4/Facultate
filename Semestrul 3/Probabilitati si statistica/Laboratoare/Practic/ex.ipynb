```python
from scipy.stats import expon, randint,uniform
from matplotlib.pyplot import hist, plot,grid,show,xticks
from numpy import linspace,mean
from math import floor

data = expon.rvs(scale=2,size=10000)
hist(data,bins=10,range=(0,10),density=True,rwidth=0.9,color='red')
xticks(range(0,11))
x=linspace(0,10,10000)
plot(x,expon.pdf(x,scale=2))
grid()
show()
```

```python
note=uniform.rvs(loc=1,scale=9,size=10000)
promovate=mean([floor(note[i])-floor(data[i])>=5 for i in range(10000)])
print(promovate)

print(1-uniform.cdf(5,loc=1,scale=9))
```
