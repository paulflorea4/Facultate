```python
import os
import random
import numpy as np
import networkx as nx
from networkx.algorithms import community
import math
```

```python
class CommunityGeneticAlgorithm:
    def __init__(self, graph, pop_size=50, generations=60, crossover_rate=0.8, mutation_rate=0.15, fitness_type='modularity'):
        self.G = graph.to_undirected()
        self.nodes = list(self.G.nodes())
        self.num_nodes = len(self.nodes)
        
        # Mapare noduri la indici 0..N-1
        self.node_to_idx = {node: i for i, node in enumerate(self.nodes)}
        self.idx_to_node = {i: node for i, node in enumerate(self.nodes)}
        
        # Structură de adiacență optimizată sub formă de indici
        self.adj_list_idx = {self.node_to_idx[u]: [self.node_to_idx[v] for v in self.G.neighbors(u)] for u in self.G.nodes()}
        
        self.pop_size = pop_size
        self.generations = generations
        self.crossover_rate = crossover_rate
        self.mutation_rate = mutation_rate
        self.fitness_type = fitness_type

    def _initialize_population(self):
        population = []
        for _ in range(self.pop_size):
            chromosome = []
            for i in range(self.num_nodes):
                if self.adj_list_idx[i]:
                    chromosome.append(random.choice(self.adj_list_idx[i]))
                else:
                    chromosome.append(i)
            population.append(chromosome)
        return population

    def _decode(self, chromosome):
        temp_graph = nx.Graph()
        temp_graph.add_nodes_from(range(self.num_nodes))
        for i, target in enumerate(chromosome):
            temp_graph.add_edge(i, target)
        
        components = list(nx.connected_components(temp_graph))
        communities = [set(self.idx_to_node[i] for i in comp) for comp in components]
        return communities

    def _calculate_fitness(self, communities):
        num_comm = len(communities)
        if num_comm <= 1:
            return 0.0
    
        # --- 1. MODULARITATE ---
        if self.fitness_type == 'modularity':
            try:
                return community.modularity(self.G, communities)
            except:
                return -1.0
    
        # --- 2. CONDUCTANȚĂ ---
        elif self.fitness_type == 'conductance':
            conductances = []
            for c in communities:
                try:
                    cond = nx.conductance(self.G, c)
                    conductances.append(cond)
                except:
                    conductances.append(1.0)
            avg_cond = np.mean(conductances) if conductances else 1.0
            raw_score = 1.0 / (1.0 + avg_cond)
    
            # Penalizare pentru comunitate mamut (>85% din noduri)
            max_comm_size = max(len(c) for c in communities)
            if max_comm_size > self.num_nodes * 0.85:
                raw_score *= 0.5
    
            # Penalizare pentru prea putine comunitati fata de marimea grafului
            min_expected = max(2, int(math.sqrt(self.num_nodes) * 0.5))
            if num_comm < min_expected:
                raw_score *= (num_comm / min_expected)
    
            return raw_score
    
        # --- 3. ACOPERIRE / COVERAGE ---
        elif self.fitness_type == 'coverage':
            total_edges = self.G.number_of_edges()
            if total_edges == 0:
                return 0.0
    
            internal_edges_count = 0
            for comm in communities:
                subgraph = self.G.subgraph(comm)
                internal_edges_count += subgraph.number_of_edges()
    
            raw_coverage = float(internal_edges_count) / total_edges
    
            # Penalizare pentru comunitate mamut (>80% din noduri)
            max_comm_size = max(len(c) for c in communities)
            if max_comm_size > self.num_nodes * 0.80:
                penalty = 1.0 - (max_comm_size / self.num_nodes)
                raw_coverage *= penalty
    
            # Penalizare pentru prea putine comunitati fata de marimea grafului
            min_expected = max(2, int(math.sqrt(self.num_nodes) * 0.5))
            if num_comm < min_expected:
                raw_coverage *= (num_comm / min_expected)
    
            return raw_coverage
    
        return -1.0

    def _crossover(self, parent1, parent2):
        if random.random() > self.crossover_rate:
            return parent1.copy(), parent2.copy()
            
        child1, child2 = [], []
        for i in range(self.num_nodes):
            if random.random() < 0.5:
                child1.append(parent1[i])
                child2.append(parent2[i])
            else:
                child1.append(parent2[i])
                child2.append(parent1[i])
        return child1, child2

    def _mutate(self, chromosome):
        for i in range(self.num_nodes):
            if random.random() < self.mutation_rate:
                if len(self.adj_list_idx[i]) > 0:
                    chromosome[i] = random.choice(self.adj_list_idx[i])
        return chromosome

    def run(self):
        population = self._initialize_population()
        best_chromosome = None
        best_fitness = -float('inf')
        
        for gen in range(self.generations):
            decoded_pop = [self._decode(chrom) for chrom in population]
            fitness_scores = [self._calculate_fitness(comm) for comm in decoded_pop]
            
            for i, score in enumerate(fitness_scores):
                if score > best_fitness:
                    best_fitness = score
                    best_chromosome = population[i].copy()
            
            new_population = []
            while len(new_population) < self.pop_size:
                candidates1 = random.sample(list(zip(population, fitness_scores)), min(3, self.pop_size))
                p1 = max(candidates1, key=lambda x: x[1])[0]
                
                candidates2 = random.sample(list(zip(population, fitness_scores)), min(3, self.pop_size))
                p2 = max(candidates2, key=lambda x: x[1])[0]
                
                c1, c2 = self._crossover(p1, p2)
                c1 = self._mutate(c1)
                c2 = self._mutate(c2)
                
                new_population.extend([c1, c2])
                
            population = new_population[:self.pop_size]
            
        return self._decode(best_chromosome), best_fitness
```

```python
def evaluate_predefined_algorithm(G):
    """Abordarea 1: Algoritm predefinit din NetworkX (Louvain)"""
    communities_list = list(community.louvain_communities(G.to_undirected()))
    mod = community.modularity(G.to_undirected(), communities_list)
    return communities_list, mod

def print_results(network_name, method_name, communities, quality_score):
    print(f"\n" + "="*70)
    print(f" REȚEA: {network_name} | METODĂ: {method_name}")
    print(f"="*70)
    print(f"-> Număr total comunități detectate: {len(communities)}")
    print(f"-> Scor de performanță (Fitness/Calitate): {quality_score:.4f}")
    
    # Maparea cerută în datele de ieșire
    node_community_mapping = {}
    for i, comm in enumerate(communities):
        for node in comm:
            node_community_mapping[node] = i
            
    sample_nodes = []
    seen_communities = set()
    for node, comm_id in node_community_mapping.items():
        if comm_id not in seen_communities:
            sample_nodes.append(node)
            seen_communities.add(comm_id)
        if len(seen_communities) >= min(5, len(communities)):
            break
    print("-> Apartenența nodurilor (Eșantion):")
    for node in sample_nodes:
        print(f"   Nod [{node}] aparține comunității -> {node_community_mapping[node]}")
```

```python
all_networks = {}
base_data_path = os.path.join("data", "real")
lab_folders = ["dolphins", "football", "karate", "krebs"]

print("[1/2] Se verifică și se încarcă fișierele obligatorii")
for folder in lab_folders:
    gml_file_path = os.path.join(base_data_path, folder, f"{folder}.gml")
    
    if os.path.exists(gml_file_path):
        try:
            all_networks[f"Lab: {folder.upper()}"] = nx.read_gml(gml_file_path)
            print(f"    [OK] Încărcat cu succes: {gml_file_path}")
        except Exception as e:
            print(f"    [EROARE] Eroare la citirea {gml_file_path}: {e}. Generăm fallback.")
            all_networks[f"Lab: {folder.upper()} (Fallback)"] = nx.erdos_renyi_graph(50, 0.1)
    else:
        print(f"    [MISING] Negăsit la calea {gml_file_path}. Generăm rețea simulată.")
        all_networks[f"Lab: {folder.upper()} (Simulat)"] = nx.erdos_renyi_graph(45, 0.15)

print("\n[2/2] Se încarcă cele 6 rețele suplimentare")
all_networks["Extra 1: Les Miserables"] = nx.les_miserables_graph()
all_networks["Extra 2: Florentine Families"] = nx.florentine_families_graph()
all_networks["Extra 3: Davis Southern Women"] = nx.projected_graph(nx.davis_southern_women_graph(), nx.davis_southern_women_graph().nodes(data=False))
all_networks["Extra 4: Barabasi-Albert Graph (Scale-Free)"] = nx.barabasi_albert_graph(60, 2)
all_networks["Extra 5: Watts-Strogatz (Small-World)"] = nx.watts_strogatz_graph(50, 4, 0.1)
all_networks["Extra 6: Erdos-Renyi Random"] = nx.erdos_renyi_graph(55, 0.08)

print(f"\nPregătit! Avem un total de {len(all_networks)} rețele gata de testat.")
```

```python
for net_name, G in all_networks.items():
    # 1. Metoda Predefinită Louvain (20 pct / rețea)
    comm_pre, mod_pre = evaluate_predefined_algorithm(G)
    print_results(net_name, "Predefinit (Louvain)", comm_pre, mod_pre)
    
    # 2. Algoritm Genetic Propriu - Modularitate (50 pct / rețea)
    ga_mod = CommunityGeneticAlgorithm(G, pop_size=30, generations=30, fitness_type='modularity')
    comm_ga, fit_ga = ga_mod.run()
    print_results(net_name, "AG Student (Modularitate)", comm_ga, fit_ga)
    
    # 3. Algoritm Genetic Propriu - BONUS 1: Conductanță
    ga_cond = CommunityGeneticAlgorithm(G, pop_size=30, generations=30, fitness_type='conductance')
    comm_cond, fit_cond = ga_cond.run()
    print_results(net_name, "AG Student (BONUS 1: Conductance)", comm_cond, fit_cond)
    
    # 4. Algoritm Genetic Propriu - BONUS 2: Acoperire (Coverage)
    ga_cov = CommunityGeneticAlgorithm(G, pop_size=30, generations=30, fitness_type='coverage')
    comm_cov, fit_cov = ga_cov.run()
    print_results(net_name, "AG Student (BONUS 2: Coverage)", comm_cov, fit_cov)
    print("\n" + "#"*70 + "\n")
```
