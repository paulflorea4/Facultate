```python
from random import sample
from random import randint
```

```python
def MontyHall(schimba_usa,nr_sim,afisare):
    contor=0;
    for _ in range(nr_sim):
        asezare_initiala=['c','c','c']
        masina,capra1,capra2=sample([0,1,2],3)
        asezare_initiala[masina]='m'
        alegere_initiala=randint(0,2)
        if alegere_initiala==masina:
            aleg_prez=sample([capra1,capra2],1).pop()
        elif alegere_initiala==capra1:
            aleg_prez=capra2
        else:
            aleg_prez=capra1
        if schimba_usa:
            aleg_finala=({0,1,2}-{alegere_initiala,aleg_prez}).pop()
        else:
            aleg_finala=alegere_initiala

        contor+=aleg_finala==masina
    return contor/nr_sim
```

```python
MontyHall(True,10000,False)
```

```python
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend
data = [randrange(1,7) for _ in range(500)]
bin_edges = [k+0.5 for k in range(8)] #1.5 2.5 2.5. 3.5 4.5 5.5 6.5
hist(data,bins=bin_edges , density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
    alpha = 0.5, label = 'frecvente relative')

distribution = dict([(i,1/6) for i in range(1,7)])
bar(list(distribution.keys()),list(distribution.values()), width = 0.7, color = 'red', edgecolor = 'black',
    alpha= 0.6, label = 'probabilitati')

legend(loc = 'lower left')
grid()
show()
```
